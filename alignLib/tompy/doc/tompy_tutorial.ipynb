# Introduction to Tompy

## What is Tompy?

Tompy is a software package in Python that provides the basic operations necessary for processing the data from cryo-electron tomography.

## What is the advantage over other tools, e.g. Tom and PyTom?

- Tompy is meant to be an alternative of Tom, since Tom toolbox is written in Matlab which is a **commercial** software.
- While the advantage compared to PyTom is that Tompy is built on top of Numpy, a highly efficient numerical computing library in Python. Therefore, the user can take full advantage of Numpy if he/she is familiar with it. **No** additional low level C/C++ function **needs to be complied or implemented** again like in PyTom, which is typically not a trivial task, time-consuming, and prone to bugs!

### Transforms:

```python
from tompy.transform import *
```

```python
# rotate a volume
v2 = rotate3d(v, 30,0,0)
```

```python
show2D(v2)
```

```python
#translate a volume
v2 = translate3d(v, 10,0,0)
```

```python
show2D(v2)
```

```python
v2 = v[25:75, 25:75, 25:75]
```

```python
show2D(v2)
```

```python
# Fourier transform
fv = fft(v)
v2 = ifft(fv)
np.allclose(v, v2, 1e-6)
```

### Comparisons between Tompy, PyTom and Tom

Experiments conducted on a volume of size $100^3$

```python
import matplotlib.pyplot as plt

tompy_time = (1.36, 98.4, 4.15, 5.78, 25.1)
pytom_time = (8.08, 101, 1.24, 4.92, 1.76)
tom_time = (20.146, 155.147, 0.72, 0.645, 12.21)

ind = np.arange(len(tompy_time))  # the x locations for the groups
width = 0.25       # the width of the bars

fig = plt.figure()
ax = fig.add_subplot(111)

rects1 = ax.bar(ind, tompy_time, width, color='r')
rects2 = ax.bar(ind+width, pytom_time, width, color='y')
rects3 = ax.bar(ind+2*width, tom_time, width, color='g')

# add some
ax.set_ylabel('Time (ms)')
ax.set_title('Comparison #1')
ax.set_xticks(ind+width)
ax.set_xticklabels( ('Read', 'Write', 'Max', 'Mean', 'Variance') )

ax.legend( (rects1[0], rects2[0], rects3[0]), ('Tompy', 'PyTom', 'Tom') )

plt.show()
```

## Finally, how to extend Tompy?

**No pain anymore!** The same way as to extend Numpy. Right now the easiest way is to use numpy.i, a swig interface which defines how you can communicate with the Numpy arrays. An example:

    %module swig_frm
    %{
    #define SWIG_FILE_WITH_INIT
    int frm(double *f, int dim1, double *g, int dim2, double *res, int dim3);
    %}
    %include "numpy.i"
    %init %{
        import_array();
    %}
    
    int frm(double *IN_ARRAY1, int DIM1, double *IN_ARRAY1, int DIM1, double *INPLACE_ARRAY1, int DIM1);

## End!
